```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn.utils import resample
```

```python
data = pd.read_csv("C:\\Users\\LENOVO\\Downloads\\Online_Payments_Fraud_Detection.csv")
print(data.head())
```

```python
print(data.info())
```

```python
data['isFraud'].value_counts()
```

```python
sns.histplot(data[data['isFraud'] == 1]['amount'], color='red', label='Fraudulent', kde=True)
sns.histplot(data[data['isFraud'] == 0]['amount'], color='blue', label='Non-Fraudulent', kde=True)
plt.title('Transaction Amount Distribution')
plt.xlabel('Transaction Amount')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
fraudulent_transactions = data[data['isFraud'] == 1]
non_fraudulent_transactions = data[data['isFraud'] == 0]

desired_non_fraudulent_length = len(fraudulent_transactions) * 9  

non_fraudulent_transactions_sampled = non_fraudulent_transactions.sample(n=desired_non_fraudulent_length, random_state=42)

balanced_data = pd.concat([fraudulent_transactions, non_fraudulent_transactions_sampled])

balanced_data = balanced_data.sample(frac=1, random_state=42).reset_index(drop=True)

print(balanced_data['isFraud'].value_counts())
```

```python
balanced_data.isnull().sum()
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
sns.countplot(x='isFraud', data=data)  
plt.title('Distribution of data (Before Subsampling)')
plt.subplot(1, 2, 2)
sns.countplot(x='isFraud', data=balanced_data)  
plt.title('Distribution of data (After Subsampling)')
plt.tight_layout()
plt.show
```

```python
numeric_balanced_data = balanced_data.select_dtypes(include=['number'])
correlation_matrix = numeric_balanced_data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
small_sample = balanced_data.sample(n=1000, random_state=42)
small_sample['isFraud'] = small_sample['isFraud'].astype(str)
sns.pairplot(small_sample, hue='isFraud', palette=['blue', 'red'])
plt.suptitle('Features by Fraud Class')
plt.show()
```

```python
balanced_data = balanced_data[['oldbalanceOrg', 'newbalanceOrig', 'oldbalanceDest', 'newbalanceDest','isFraud']]
for col in balanced_data:
   print(balanced_data[col].value_counts()) 
```

```python
print(balanced_data.columns)
```

```python
balanced_data.rename(columns={
    'oldbalanceDest': 'Old Balance Destination',
    'newbalanceDest': 'New Balance Destination',
    'oldbalanceOrg': 'Old Balance Sender',
    'newbalanceOrig': 'New Balance Sender'
}, inplace=True)

print(balanced_data.head())
```

```python
X1 = balanced_data.drop('isFraud',axis=1)
y1 = balanced_data['isFraud']
X2 = data.drop('isFraud',axis=1)
y2 = data['isFraud']
X1_train,X1_test,y1_train,y1_test = train_test_split(X1,y1,test_size=0.30,random_state=1)
X2_train,X2_test,y2_train,y2_test = train_test_split(X2,y2,test_size=0.30,random_state=1)
```

```python
classifier1 = LogisticRegression (solver='liblinear',max_iter=1000)
classifier1.fit(X1_train,y1_train)
columns_to_drop = ['type', 'nameOrig', 'nameDest']
X2_train_final = X2_train.drop(columns=columns_to_drop)
classifier2 = LogisticRegression (solver='liblinear',max_iter=1000)
classifier2.fit(X2_train_final, y2_train)
```

```python
from sklearn.model_selection import RandomizedSearchCV, train_test_split
from scipy.stats import randint
param_dist = {
    'n_estimators': randint(100, 500),  
    'max_depth': [None] + list(randint(1, 50).rvs(5)),  
    'min_samples_split': randint(2, 10),  
    'min_samples_leaf': randint(1, 10) 
}

rf_classifier1 = RandomForestClassifier()
rf_classifier2 = RandomForestClassifier()

random_search1 = RandomizedSearchCV(estimator=rf_classifier1, param_distributions=param_dist, n_iter=20,
                                    cv=5, scoring='accuracy', random_state=1, n_jobs=-1)
random_search1.fit(X1_train, y1_train)

random_search2 = RandomizedSearchCV(estimator=rf_classifier2, param_distributions=param_dist, n_iter=20,
                                    cv=5, scoring='accuracy', random_state=1, n_jobs=-1)
random_search2.fit(X2_train, y2_train)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Predictions for logistic regression and random forest on balanced data
log_reg_pred1 = classifier1.predict(X1_test)
rf_pred1 = random_search1.predict(X1_test)

# Predictions for logistic regression and random forest on unbalanced data
log_reg_pred2 = classifier2.predict(X2_test)
rf_pred2 = random_search2.predict(X2_test)

# Metrics for logistic regression on balanced data
log_reg_accuracy1 = accuracy_score(y1_test, log_reg_pred1)
log_reg_precision1 = precision_score(y1_test, log_reg_pred1)
log_reg_recall1 = recall_score(y1_test, log_reg_pred1)
log_reg_f1_1 = f1_score(y1_test, log_reg_pred1)

# Metrics for logistic regression on unbalanced data
log_reg_accuracy2 = accuracy_score(y2_test, log_reg_pred2)
log_reg_precision2 = precision_score(y2_test, log_reg_pred2)
log_reg_recall2 = recall_score(y2_test, log_reg_pred2)
log_reg_f1_2 = f1_score(y2_test, log_reg_pred2)

# Metrics for random forest on balanced data
rf_accuracy1 = accuracy_score(y1_test, rf_pred1)
rf_precision1 = precision_score(y1_test, rf_pred1)
rf_recall1 = recall_score(y1_test, rf_pred1)
rf_f1_1 = f1_score(y1_test, rf_pred1)

# Metrics for random forest on unbalanced data
rf_accuracy2 = accuracy_score(y2_test, rf_pred2)
rf_precision2 = precision_score(y2_test, rf_pred2)
rf_recall2 = recall_score(y2_test, rf_pred2)
rf_f1_2 = f1_score(y2_test, rf_pred2)

# Print results
print("Logistic Regression on balanced data:")
print("Accuracy:", log_reg_accuracy1)
print("Precision:", log_reg_precision1)
print("Recall:", log_reg_recall1)
print("F1 Score:", log_reg_f1_1)
print("\n")

print("Logistic Regression on unbalanced data:")
print("Accuracy:", log_reg_accuracy2)
print("Precision:", log_reg_precision2)
print("Recall:", log_reg_recall2)
print("F1 Score:", log_reg_f1_2)
print("\n")

print("Random Forest on balanced data:")
print("Accuracy:", rf_accuracy1)
print("Precision:", rf_precision1)
print("Recall:", rf_recall1)
print("F1 Score:", rf_f1_1)
print("\n")

print("Random Forest on unbalanced data:")
print("Accuracy:", rf_accuracy2)
print("Precision:", rf_precision2)
print("Recall:", rf_recall2)
print("F1 Score:", rf_f1_2)
```
